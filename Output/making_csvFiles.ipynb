```python
# getting win rates
champion_win_dict=dict()
with open('/content/drive/MyDrive/BigDataMiniProject2/Output/Result/champion_vs_wins') as win:
  for line in win:
    champion_win=line.replace('(','').replace(')','').strip().split(',')
    champion_win_dict[champion_win[0][1:-1]]=int(champion_win[1])
champion_pick_dict=dict()
with open('/content/drive/MyDrive/BigDataMiniProject2/Output/Result/champion_vs_times') as pick:
  for line in pick:
    champion_pick=line.replace('(','').replace(')','').strip().split(',')
    champion_pick_dict[champion_pick[0][1:-1]]=int(champion_pick[1])
#_win_rate=dict()
#for champion in champion_pick_dict:
#  champion_win_rate[champion]= (champion_win_dict[champion]/champion_pick_dict[champion])* 100

#champion_tuples=[("{}".format(i),champion_pick_dict[i],champion_win_dict[i]) for i in champion_pick_dict if i in champion_win_dict]

```

```python
champion_vs_winRate=dict()
for i in champion_win_dict:
  champions="**".join(list(sorted(list(i.split('**')))))
  picks=champion_pick_dict[champions]
  wins=champion_win_dict[i]
  champion_vs_winRate[i]=(picks,wins,(wins/picks)*100)
```

```python
champion_won=[i.split("**")[0] for i in champion_vs_winRate.keys()]
champion_lost=[i.split("**")[1] for i in champion_vs_winRate.keys()]
picks_values=[i[0] for i in champion_vs_winRate.values()]
wins_values=[i[1] for i in champion_vs_winRate.values()]
rates_values=[i[2] for i in champion_vs_winRate.values()]
```

```python
import pandas as pd

df=pd.DataFrame({'Champion':champion_won,'Against':champion_lost,'VS. Times':picks_values, 'VS. Won':wins_values,'VS. Win Rate':rates_values})
df
```

```python
df['win_rate']=(df['win']/df['pick'])*100
```

```python
df['pick_rate']=(df['pick']/sum(df['pick']))*100
```

```python
df['score']=(df['win_rate'])*df['pick_rate']
```

```python
df.to_csv('champions_VS_rates.csv')
```

```python
import json
with open('/content/drive/MyDrive/BigDataMiniProject2/DATADRAGON/championFull.json') as file:
  champion_datajson=json.load(file)
  champions = champion_datajson["data"].keys()
```

```python
len(champions)
```

```python
champion_to_best_worst=dict()
for champion in champions:
  against_champion_rate=[]
  exist=False
  for key in champion_vs_winRate:
    current=list(key.split('**'))[0]
    against=list(key.split('**'))[1]
    if current==champion:
      exist=True
      rate=champion_vs_winRate[key][2]
      against_champion_rate.append((against,rate))
  if exist==True:
    best=max(against_champion_rate,key=lambda k:k[1])
    worst=min(against_champion_rate,key=lambda k:k[1])
    champion_to_best_worst[champion]=(best,worst)
    
```

```python
champion_to_best_worst
```

```python
championNames=champion_to_best_worst.keys()
bestchampions=[i[0] for i in champion_to_best_worst.values()]
worstchampions=[i[1] for i in champion_to_best_worst.values()]
df_against=pd.DataFrame({'Champion':championNames,'Best Enemy, winRate':bestchampions,'Worst Enemy, winRate':worstchampions})
df_against
```
